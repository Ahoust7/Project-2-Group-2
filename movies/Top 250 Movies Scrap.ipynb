```python
from splinter import Browser
from bs4 import BeautifulSoup as soup
```

```python
# Set up Splinter
browser = Browser('chrome')
```

```python
# Visit IMDB
url = 'https://www.imdb.com/chart/top/?ref_=nv_mv_250'
browser.visit(url)
```

```python
# Parse the website
html = browser.html
html_soup = soup(html, 'html.parser')

print(html_soup)
```

```python
# Find the list of movies
movie_list = html_soup.find('tbody', {'class': 'lister-list'})

# Find all the rows in the list
movie_rows = movie_list.find_all('tr')
```

```python
# Loop over the rows and extract the movie details
for row in movie_rows:
    # Get the movie title
    title = row.find('td', {'class': 'titleColumn'}).find('a').text

    # Get the movie year
    year = row.find('td', {'class': 'titleColumn'}).find('span', {'class': 'secondaryInfo'}).text.strip('()')

    # Get the movie rating
    rating = row.find('td', {'class': 'ratingColumn'}).find('strong').text

    # Print the movie details
    print(f"{title} ({year}) - {rating}")
```
