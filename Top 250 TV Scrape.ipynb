```python
from splinter import Browser
from bs4 import BeautifulSoup as soup
```

```python
# Set up Splinter
browser = Browser('chrome')
```

```python
# Visit IMDB
url = 'https://www.imdb.com/chart/toptv/?ref_=nv_tvv_250'
browser.visit(url)
```

```python
# Parse the website
html = browser.html
html_soup = soup(html, 'html.parser')
```

```python
# Find the list of movies
TV_list = html_soup.find('tbody', {'class': 'lister-list'})

# Find all the rows in the list
TV_rows = TV_list.find_all('tr')
```

```python
# Loop over the rows and extract the TV details
for row in TV_rows:
    # Get the TV title
    title = row.find('td', {'class': 'titleColumn'}).find('a').text

    # Get the TV year
    year = row.find('td', {'class': 'titleColumn'}).find('span', {'class': 'secondaryInfo'}).text.strip('()')

    # Get the TV rating
    rating = row.find('td', {'class': 'ratingColumn'}).find('strong').text

    # Print the movie details
    print(f"{title} ({year}) - {rating}")
```

```python
import pandas as pd
```

```python
# Create an empty list to store the movie data
TV = []
```

```python
# Find the list of TV
TV_list = html_soup.find('tbody', {'class': 'lister-list'})
```

```python
# Find all the rows in the list
TV_rows = TV_list.find_all('tr')
```

```python
# Loop over the rows and extract the movie details
for row in TV_rows:
    # Get the movie title
    title = row.find('td', {'class': 'titleColumn'}).find('a').text

    # Get the movie year
    year = row.find('td', {'class': 'titleColumn'}).find('span', {'class': 'secondaryInfo'}).text.strip('()')

    # Get the movie rating
    rating = row.find('td', {'class': 'ratingColumn'}).find('strong').text

    # Add the movie data to the list
    TV.append({'Title': title, 'Year': year, 'Rating': rating})
```

```python
# Create a DataFrame from the movie data
df = pd.DataFrame(TV)
```

```python
print(df)
```

```python
# Export the DataFrame to a CSV file
df.to_csv('imdb_top_250_TV.csv', index=False)
```
