This notebook reads the Top 1000 movies csv file, transformsand cleans the data, and spits out a JSON file

```python
# Import dependencies
import pandas as pd
import csv
import json
import os
import ast
import warnings
warnings.filterwarnings('ignore')
```

```python
# Read CSV
df = pd.read_csv('/Users/meganconnelly/Project-2-Group-2/imdb_top_1000_movies_final.csv')
```

```python
# Show CSV
df.head()
```

```python
# Convert the string representation of the list to an actual list
df['Stars'] = df['Stars'].apply(lambda x: ast.literal_eval(x))
df['Genres'] = df['Genres'].apply(lambda x: ast.literal_eval(x))
```

```python
# Clean up the 'Year' column by removing parentheses and Roman numerals I, II, and III
df['Year'] = df['Year'].str.replace('(', '').str.replace(')', '').str.replace('I', '').str.replace('II', '').str.replace('III', '')
```

```python
# Show the dataframe
df
```

```python
# Drop the 'Gross' column
df = df.drop('Gross', axis=1)
```

```python
# Show the dataframe
df
```

```python
# Export dataframe to JSON file
df.to_json('imdb_top_1000_movies.json', orient='records')
```
